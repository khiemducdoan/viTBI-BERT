```python
import numpy as np
import matplotlib.pyplot as plt

# Data
datasets = ['PostMN_TBI_MH', 'PreMN_TBI_MH', 'MN_TBI_MH']
models = ['PhoBERTbase', 'ViPubmedDeBERTaxsmall', 'ViHealthBERTbase-word']
accuracies = [
    [0.59, 0.58, 0.65],  # PostMN_TBI_MPH
    [0.65, 0.62, 0.68],  # PreMN_TBI_MPH
    [0.68, 0.68, 0.71],  # MN_TBI_MPH
]

# Parameters
bar_width = 0.25
index = np.arange(len(datasets))

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Create bars for each model
for i, model in enumerate(models):
    plt.bar(index + i * bar_width, [acc[i] for acc in accuracies], bar_width, label=model)

# Adding labels and title
plt.xlabel('Datasets', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Model Accuracies Across Different Datasets', fontsize=14)
plt.xticks(index + bar_width, datasets)
plt.ylim(0.5, 0.8)
plt.legend()

# Adding accuracy values on top of bars
for i in range(len(datasets)):
    for j in range(len(models)):
        plt.text(index[i] + j * bar_width, accuracies[i][j] + 0.01, f'{accuracies[i][j]:.2f}', ha='center', va='bottom')

# Display the plot
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
